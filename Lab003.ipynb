```python
def is_par(n):
    return n % 2 == 0
```

```python
import unittest

class TestIsPar(unittest.TestCase):
    def test_par(self):
        self.assertTrue(is_par(2))

    def test_impar(self):
        self.assertFalse(is_par(3))

    def test_zero(self):
        self.assertTrue(is_par(0))

    def test_negativo(self):
        self.assertTrue(is_par(-4))

unittest.main(argv=[''], exit=False)
```

```python
def fatorial(n):
    if n < 0:
        raise ValueError("n deve ser >= 0")
    if n == 0:
        return 1
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado
```

```python
import unittest

class TestFatorial(unittest.TestCase):
    def test_zero(self):
        self.assertEqual(fatorial(0), 1)

    def test_cinco(self):
        self.assertEqual(fatorial(5), 120)

    def test_negativo(self):
        with self.assertRaises(ValueError):
            fatorial(-1)

unittest.main(argv=[''], exit=False)
```

```python
class InsufficientFunds(Exception):
    pass

class Conta:
    def __init__(self, saldo=0):
        self.saldo = saldo

    def depositar(self, amount):
        if amount < 0:
            raise ValueError("Depósito não pode ser negativo")
        self.saldo += amount

    def sacar(self, amount):
        if amount < 0:
            raise ValueError("Saque não pode ser negativo")
        if amount > self.saldo:
            raise InsufficientFunds("Saldo insuficiente")
        self.saldo -= amount
```

```python
import unittest

class TestConta(unittest.TestCase):
    def setUp(self):
        self.conta = Conta(100)

    def test_deposito_valido(self):
        self.conta.depositar(50)
        self.assertEqual(self.conta.saldo, 150)

    def test_deposito_invalido(self):
        with self.assertRaises(ValueError):
            self.conta.depositar(-10)

    def test_saque_sucesso(self):
        self.conta.sacar(30)
        self.assertEqual(self.conta.saldo, 70)

    def test_saque_insuficiente(self):
        with self.assertRaises(InsufficientFunds):
            self.conta.sacar(200)

    def test_saque_negativo(self):
        with self.assertRaises(ValueError):
            self.conta.sacar(-5)

unittest.main(argv=[''], exit=False)
```

```python
import requests

def buscar_clima(cidade):
    url = f'https://api.exemplo/clima?cidade={cidade}'
    response = requests.get(url)
    data = response.json()
    if 'temperatura' not in data:
        raise KeyError("Resposta inválida da API: falta 'temperatura'")
    return data['temperatura']
```

```python
import unittest
from unittest.mock import patch, Mock

class TestBuscarClima(unittest.TestCase):

    @patch('requests.get')
    def test_clima_sucesso(self, mock_get):
        mock_response = Mock()
        mock_response.json.return_value = {'temperatura': 25}
        mock_get.return_value = mock_response

        resultado = buscar_clima("São Paulo")
        self.assertEqual(resultado, 25)

    @patch('requests.get')
    def test_clima_sem_temperatura(self, mock_get):
        mock_response = Mock()
        mock_response.json.return_value = {'umidade': 80}
        mock_get.return_value = mock_response

        with self.assertRaises(KeyError):
            buscar_clima("São Paulo")

    @patch('requests.get')
    def test_excecao_requests(self, mock_get):
        mock_get.side_effect = requests.exceptions.RequestException("Erro de rede")
        with self.assertRaises(requests.exceptions.RequestException):
            buscar_clima("Rio de Janeiro")

unittest.main(argv=[''], exit=False)
```
