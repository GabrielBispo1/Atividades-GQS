```python
#Função maior
def maior(a, b):
    return a if a > b else b
```

```python
#Testes
def test_maior_errado():
    assert maior(5, 3) == 3  # Falha proposital

def test_maior_forcado():
    assert maior(5, 5) == 5  # Forçado

def test_maior_correto():
    assert maior(10, 7) == 10  # Correto

#test_maior_errado()
test_maior_forcado()
test_maior_correto()
```

```python
#Função string invertida
def inverter_string(texto: str) -> str:
    return texto[::-1]
```

```python
#Testes
def test_inverter_errado():
    assert inverter_string("abc") == "abc"  # Falha proposital

def test_inverter_forcado():
    assert inverter_string("x") == "x"  # Forçado

def test_inverter_correto():
    assert inverter_string("python") == "nohtyp"  # Correto
```

```python
#Função validar cpf
def valida_cpf(cpf: str) -> bool:
    return cpf.isdigit() and len(cpf) == 11
```

```python
#Testes
def test_valida_cpf_errado():
    assert valida_cpf("12345678901") is False  # Falha proposital

def test_valida_cpf_forcado():
    assert valida_cpf("00000000000") is True  # Forçado

def test_valida_cpf_correto():
    assert valida_cpf("12345") is False  # Correto
```
